## Thống kê sự tương quan giữa các cột dữ liệu

**- Sự tương quan giữa tỷ lệ hoàn thành khóa học và tổng số lượng comments của học viên cho khóa học.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# Sample data based on the image
data = {
    'Total comments': ['0-10', '11-100', '101-1000', '1001-10000', '10001-70000'],
    'Number of courses': [3000, 500, 300, 100, 50],
    'Average course completion rate': [0.65, 0.7, 0.75, 0.8, 0.85]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Set the x locations and bar width
x = np.arange(len(df['Total comments']))
width = 0.4

fig, ax1 = plt.subplots(figsize=(10, 6))

# Bar plot for Number of courses
ax1.bar(x, df['Number of courses'], color=['#3e3b7f', '#7f9bae', '#8fc0c1', '#a5d3c8', '#c9e6e3'], width=width)
ax1.set_xlabel('Total comments')
ax1.set_ylabel('Number of courses', color='black')

# Line plot for Average course completion rate
ax2 = ax1.twinx()
ax2.plot(x, df['Average course completion rate'], color='orange', marker='o', label='Average course completion rate')
ax2.fill_between(x, np.array(df['Average course completion rate']) - 0.05, np.array(df['Average course completion rate']) + 0.05, color='orange', alpha=0.2)
ax2.set_ylabel('Average course completion rate', color='black')

# Set x-ticks
ax1.set_xticks(x)
ax1.set_xticklabels(df['Total comments'])

# Add legend
ax2.legend(loc='upper left')

plt.title('Total comments vs Number of courses and Completion Rate')
plt.show()
```

**Nhận xét:** Biểu đồ cho thấy mối quan hệ giữa tổng số bình luận và số lượng khóa học, cùng với tỷ lệ hoàn thành khóa học trung bình. Khi tổng số bình luận tăng, số lượng khóa học giảm nhưng tỷ lệ hoàn thành khóa học trung bình lại tăng, đạt gần 0.85 khi có từ 10,001 đến 70,000 bình luận. Điều này cho thấy rằng các khóa học có nhiều bình luận hơn có tỷ lệ hoàn thành cao hơn.

**- So sánh sự phân bố về tỉ lệ hoàn thành khóa học giữa các khóa học có bình luận và không có bình luận**

```python
bins = list(range(0, int(df['rank'].max()) + 1000, 100))
labels = list(range(1, len(bins)))

df['rank_binned'] = pd.cut(df['rank'], bins=bins, labels=labels, right=False)
```

```python
df
```

```python
# Dùng trong xử lý dữ liệu
df['comment_ratio'] = np.where(
    (df['total_cmt'].notnull()) & (df['positive'] != 0),
    df['negative'] / df['positive'],
    0  # Assign 0 if `negative` is 0 or `total_cmt` is null
)
```

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()

df[['num_users_normalized', 'average_completion_rate_normalized', 'resource_normalized','rank_normalized']] = scaler.fit_transform(
    df[['num_users', 'average_completion_rate', 'number of resources','rank_binned']]
)
```

```python
# Mô tả các cột 'good_comment_ratio', 'bad_comment_ratio', 'num_users_normalized', 'average_completion_rate_normalized', 'resource_normalized', 'total_cmt', 'positive', 'negative', và 'neutral'
df[['comment_ratio', 'num_users_normalized', 'average_completion_rate_normalized', 'resource_normalized', 'total_cmt', 'positive', 'negative', 'neutral','rank_normalized']].describe()
```

```python
import pandas as pd


w1, w2, w3, w4, w5 = 0.35, 0.2, 0.15, -0.05, 0.35

def adjust_comment_ratio(comment_ratio):
    return comment_ratio * (1 if comment_ratio < 1 else -1)

adjusted_comment_ratio = df['comment_ratio'].apply(adjust_comment_ratio)

df['composite_score'] = (
    w1 * df['num_users_normalized'] +
    w2 * df['average_completion_rate_normalized'] +
    w3 * adjusted_comment_ratio +
    w4 * df['rank_normalized'] +
    w5 * df['resource_normalized']
)

```

```python
# Define thresholds for five categories (e.g., very bad, bad, normal, good, very good)
quantiles = df['composite_score'].quantile([0.2, 0.4, 0.6, 0.8])

# Classify courses into five categories based on composite_score
df['course_classification'] = pd.cut(
    df['composite_score'],
    bins=[-float('inf'), quantiles.iloc[0], quantiles.iloc[1], quantiles.iloc[2], quantiles.iloc[3], float('inf')],
    labels=['Very dissatisfied', 'Dissatisfied', 'Neutral', 'Satisfied', 'Very satisfied']
)
```

## Khái phá dữ liệu

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split

# Read data from CSV file
df = pd.read_csv('final_dataset.csv')  # Replace with the actual file path

# Drop rows with NaN values in any column
df = df.dropna()

# Prepare feature matrix X and target vector y
X = df.drop(columns=['id', 'name', 'course_classification'])
y = df['course_classification']

# Encode target variable if it's categorical
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Standardize features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets with an 80/20 ratio
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)

# Train the Random Forest Classifier model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Calculate feature importances
importances = rf_model.feature_importances_

# Create a Series for easy sorting and display
forest_importances = pd.Series(importances, index=X.columns)
sorted_importances = forest_importances.sort_values(ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sorted_importances.plot(kind='barh', color='blue')
plt.xlabel("Feature Importance")
plt.ylabel("Features")
plt.title("Feature Importance in Random Forest Classifier")
plt.gca().invert_yaxis()  # Invert y-axis to show the most important features at the top
plt.show()
```
